#### 3, 4 y 5
Aplicar regresión logística con entropía cruzada como función de costo, crear una función que recibe de entrada un tensor x y predice la clase(utilizar el diccionario para mostrar el texto perteneciente al label).

Adicional a el costo(por entropía cruzada) en el grafo se debe agregar un nodo que calcule el acurracy(comparando el valor real vs el estimado por el modelo).

El accuracy y el costo deben ser agregados a tensorboard (agregar screenshots al notebook)


````python
class LogisticRegression:
    def __init__(self,pixels,clases):
        #tf.reset_default_graph()
        
        self.b = tf.Variable(tf.truncated_normal([pixels * pixels, clases]))

        self.biases = tf.Variable(tf.zeros([clases]))
        self.error = tf.get_variable("error_total",dtype=tf.float32,
                                  shape=[],initializer=tf.zeros_initializer())
        self.prediccion = tf.get_variable("accuracy_total",dtype=tf.float32,
                                  shape=[],initializer=tf.zeros_initializer())

        
    def step(self,x,y,lr):
        # hipotesis
        yhat = tf.add(tf.matmul(x, self.b, name="matmul"), self.biases,name="hipotesis")
        
        error = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits = yhat,labels= y, name="sfmax_entropy"),name="total_entropy")
        
        # Optimizacion
        # optimizer = tf.train.GradientDescentOptimizer(0.5).minimize(loss)
        grad = tf.gradients(error,[self.b,self.biases], name = "gradients")
        
        local_bn = tf.assign(self.b, self.b - tf.multiply(lr,grad[0],name="multi_lr_grad"), name="assign_beta")
        local_bias = tf.assign(self.biases, self.biases - tf.multiply(lr,grad[1],name="multi_lr_grad_bias"), name="assign_beta")
        local_error = tf.assign(self.error, error , name="assign_error")

        with tf.name_scope("accuracy") as scope:
            type_yhat = tf.math.argmax(yhat, 1, name="type_yhat")
            type_y = tf.math.argmax(y, 1, name="type_y")
            cant_ok = tf.reduce_sum(tf.cast(tf.math.equal(type_yhat,type_y), tf.float32), name = "correctos")
            total = tf.cast(tf.shape(y)[0],tf.float32, name = "total")
            local_pred = tf.assign(self.prediccion, (100.0 * cant_ok / total) , name = "asig_accuracy")
        #local_pred = tf.assign(self.prediccion,tf.shape(y))
        salida = tf.group(local_bn,local_bias,local_error, local_pred, name="grupo")

        return salida
````

````python
def playModel(xdat,ydat,image_size=28 ,cntClass=3, lr = 0.002,status = 50,batch_size = 32,epochs = 151,name="" ):
    tf.reset_default_graph()

    largo = len(v_filter_train_images)
    iteraciones = int(largo/batch_size)
    gd = LogisticRegression(image_size,cntClass)

    dat_x = tf.placeholder(tf.float32,[None,image_size*image_size],"Value_x")
    dat_y = tf.placeholder(tf.float32,[None,cntClass],"Value_y")

    step = gd.step(dat_x,dat_y,lr)

    error_summary = tf.summary.scalar('error_sum', gd.error)
    accu_summary = tf.summary.scalar('accuracy_sum', gd.prediccion)

    with tf.train.MonitoredSession() as session:
        localdir = "logs_tarea4/LogReg_" + datetime.now().strftime("%Y%m%d-%H%M%S") + "_" + name + "_lr=" + str(lr) + "_batch="+ str(batch_size) + "_ephocs=" + str(epochs)
        writer = tf.summary.FileWriter(localdir , session.graph)

        for i in range(epochs):

            for j in range(iteraciones):
                inicio = j * batch_size
                fin = inicio + batch_size
                feed = {dat_x:xdat[inicio:fin], dat_y:ydat[inicio:fin]}
                session.run(step,feed_dict=feed)
            esummary ,asummary = session.run([error_summary,accu_summary],feed_dict=feed)
            writer.add_summary(esummary, i)
            writer.add_summary(asummary, i)
                
            #feed = {dat_x:v_filter_train_images, dat_y:labels_hot}
            #session.run(step,feed_dict=feed)
            #summary = session.run(error_summary,feed_dict=feed)
            if i%status ==0:
                error,pred = session.run([gd.error,gd.prediccion],feed_dict=feed)
                print('epochs',i)
                print("error = {}; accuracy = {}".format(error,pred))


        b,bias,error,pred = session.run([gd.b,gd.biases,gd.error,gd.prediccion],feed_dict=feed)
        print("y = {} {}".format(b,bias))
        print("error = {}; accuracy = {}".format(error,pred))
    return (b,bias)
````

#### 8
Basado en los ejemplos de visualización de imagenes en el notebook ,crear una función que reciba de entrada el tensor que representa a una imagen, grafique esta imagen poniendo como título el nombre de la prenda con mayor probabilidad.

````python
def mostrar(data,y_orig,b,bias):
    locx = data.reshape(1,784) #.reshape(784,1)

    # calcula los valores (logits) de cada clase
    y_teorico = np.matmul(locx, orig_b) + bias
    
    # a los logits se pasan por argmax y se selecciona el indice con mayor valor
    seleccion = np.argmax(softmax(y_teorico))
    # se le suma 7 por que los valores seleccionados del catalogon son 7,8 y 9
    # a partir de un vector de 0, 1 y 2
    print("Prediccion: {} ".format(clases[seleccion + 7] ))
    
    # se muestra la categoria del elemento seleccionado
    print("Original: {} ".format(clases[y_orig]))
    
    # muestra de imagen
    plt.imshow(data, cmap = 'gray')
    plt.colorbar()
    plt.show()
````

#### Accuracy

Al analizar el Accuracy; esta correlaciona lo mismo que se logra ver con el error.  Y es que el modelo con los datos normales logra un accuracy del 100% en las iteraciones que le toma al error llegar a 0.  Pero esto no se consiguie con los otros modelos


<img src="Tarea4_accuracy.png">

El hecho de lograr un **error = 0; accuracy=100%** se puede interpretar como un overfitting del modelo; lo que sabemos que no es bueno. Por lo que se decidio probar distintos modelos para contar cuantas fallas genera; y resulto que al utilizar la data de prueba generan los mismos errores. Arriba del 90% de Accuracy generan los mismos errores por lo que no hay diferencia significativa

````python
def cont_dif(data,y_orig,b,bias):
    locx = data.reshape(1,784) #.reshape(784,1)
    # calcula los valores (logits) de cada clase
    y_teorico = np.matmul(locx, orig_b) + bias
    # a los logits se pasan por argmax y se selecciona el indice con mayor valor
    seleccion = np.argmax(softmax(y_teorico))
    if seleccion + 7 != y_orig:
        return 1
        #print("    " ,seleccion + 7, y_orig)
    
    return 0
````

````python
# Compracion de modelos
# cuantas fallas tienen en corridas aleatorias

contNormal = 0
contReg = 0
for i in range(100):
    seleccion = random.randint(0,2999)
    #print(seleccion)
    contNormal += cont_dif(filter_test_images[seleccion],filter_test_labels [seleccion],orig_b,original_bias)
    contReg += cont_dif(filter_test_images[seleccion],filter_test_labels [seleccion],lnlNorm_b, lnlNorm_bias)
print (contNormal,contReg)
````
