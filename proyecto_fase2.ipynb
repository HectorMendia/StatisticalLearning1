##### Desition Trees

```python
class DTC_predict:
    def __init__(self,feature_names, modelo = None):
        self.modelo = modelo
        self.feature_names = feature_names
            
    def predict(self, data):
        
        return self.modelo.predict(data[self.feature_names])
```

```python
mDTC = load('dt_20200629-213753_maxDepth=100_splitter=best.joblib') 
fn_dt = ['Age','SibSp','Parch','Fare','Embarked','passenger_class','passenger_sex']
modelo_DT = DTC_predict(fn_dt, mDTC)
```
